## function for getting tweets and dump them into json file

```python
def get_tweets(username, creds): 
          
        # Authorization to consumer key and consumer secret 
        auth = tweepy.OAuthHandler(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET']) 
  
        # Access to user's access key and access secret 
        auth.set_access_token(creds['ACCESS_TOKEN'], creds['ACCESS_SECRET']) 
  
        # Calling api 
        api = tweepy.API(auth) 
        
        # Empty Array 
#         tmp=[]  
  
        tweet_list = []
    
        # getting tweet information using tweey api cursor...
        for tweet in tweepy.Cursor(api.user_timeline, id=username, tweet_mode='extended').items():
            tweet_list.append(tweet._json)
#             print('-----------------------------------------------------------------------------')
#             d = {}
#             d['text'] = tweet.full_text
#             d['created_at'] = str(tweet.created_at)
#             d['retweet_count'] = tweet.retweet_count
#             d['favorite_count'] = tweet.favorite_count
            
#             no_of_images = 0
#             for media in tweet.entities.get("media",[{}]):
#                 if media.get("type",None) == "photo":
#                     no_of_images+=1
#             if no_of_images!=0:
#                 d['no_of_images'] = no_of_images
#             else:
#                 d['no_of_images'] = None
#             tmp.append(d)
#         json_dict = {}
#         json_dict['data'] = tmp
        with open("dump.json", "w") as file:  
            json.dump(tweet_list, file)
```

## reading dump json file

```python
with open('dump.json', 'r') as file:
    tweet_list = json.load(file)
```

## parsing json file

```python
d = []
for tweet in tweet_list:
    tweet_dict = {}
    created_at = tweet['created_at']
    text = tweet['full_text']
    retweet_count = tweet['retweet_count']
    favorite_count = tweet['favorite_count']
    n = 0
    if tweet.get('entities') is not None:
        if tweet['entities'].get('media') is not None:
            media_list = tweet['entities']['media']
            for media in media_list:
                if media.get('type') is not None:
                    if media['type'] == 'photo':
                        n+=1
    if n==0:
        n = 'None'
    tweet_dict['created_at'] = created_at
    tweet_dict['text'] = text
    tweet_dict['retweet_count'] = retweet_count
    tweet_dict['favorite_count'] = favorite_count
    tweet_dict['no_of_images'] = n
    d.append(tweet_dict)
```

## converting json into dataframe

```python
dataframe = pd.DataFrame(d)
```
